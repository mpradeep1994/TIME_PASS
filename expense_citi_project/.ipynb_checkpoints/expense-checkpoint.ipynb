```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
df = pd.read_csv('C:\Users\Pradeep\Google Drive\expense_citi_project/CHK_166.csv',names=['STATUS', 'DATE', '33', 'I/O', 'DESC','AMOUNT','_','Acc','_1' ]).fillna('')
```

```python
df
```

```python
df=df.drop(df.columns[[2, 6, 8]], axis=1)
```

```python
expense_type=[]
for i in df['AMOUNT']:
    if float(i.replace(',',''))>0:
        expense_type.append('credit')
    else:
        expense_type.append('debit')
```

```python
df['expense_type'] =expense_type
```

```python
expense=[]
for i in df['AMOUNT']:
    i=i.replace('-','')
    expense.append(float(i.replace(',','')))
```

```python
df['AMOUNT'] =expense
```

```python
# All features
all_features = df.columns.tolist()
print all_features
```

```python
expense_dic={
            'credit':[],
            'debit':[]
            }
for i in range(len(df)):
    expense_dic[df['expense_type'][i]].append(round(df['AMOUNT'][i],2))
```

```python
credit=sum(expense_dic['credit'])
debit =sum(expense_dic['debit'])
```

```python
a=['credit','debit']
b=[credit,debit]
plt.bar(a,b,align='center', alpha=1)
plt.ylabel('Amount')
plt.title('Expense balance')
plt.show()
print ('Expense',debit)
print ('Income',credit)
print ('Remaing',abs(debit-credit))
```

```python
income_words=[]
expense_words=[]
for i in range(len(df)):    
    t=df['DESC'][i].split(' ')
    result = ''.join([c for c in t if c.isalpha()])
    if(df['expense_type'][i] == 'credit'):
        income_words.append(result)
    else:
        expense_words.append(result)

wordcloud(" ".join(str(x) for x in income_words),'income_words')
wordcloud(" ".join(str(x) for x in expense_words),'expense_words')
```

```python
def wordcloud(a,title):
    wordcloud = WordCloud(width = 1000, height = 500).generate(a )
    plt.figure(figsize=(20,8))
    plt.title(title)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
c_df=df.ix[(df['expense_type']=='credit' )]
d_df=df.ix[(df['expense_type']=='debit' )]

c_df=c_df.reset_index(drop=True)
d_df=d_df.reset_index(drop=True)
```

```python
desc=[]
for i in range(len(c_df)):    
    t=df['DESC'][i].split()
    desc.append( ' '.join([c for c in t if c.isalpha()]))
Counter(desc)
```

```python
c_category ={'Square':'SQUARE CASH',
             'SQC':'SQUARE CASH',
             'FED':'INTERNSHIP',
             'ONLINE':'PARENTS',
             'DISCOVER':'cash back',
             'GOOGLE':'Google',
             'ATM':'PARENTS',
             'AMAZON':'Others',
             'ZIPCAR':'Others',
             'WIRE':'PARENTS',
             'ILLINOIS':'INTERNSHIP'
    
}  
```

```python
c_df['category']='Others'
cat=c_category.keys()
for i in range(len(c_df)):
    line= c_df['DESC'][i]
    for c in cat:
        if c in line:
            c_df['category'][i]=c_category[c]
```

```python
cost={}
for i in range(len(c_df)):
    #print (c_df['category'][i]) ,  (float(c_df['AMOUNT'][i])+1)
    if c_df['category'][i] in cost:
        cost[c_df['category'][i]]= float(cost[c_df['category'][i]]) + float(c_df['AMOUNT'][i])
    else:
        cost[c_df['category'][i]]= float(c_df['AMOUNT'][i])
        
```

```python
cost
```

```python
explode = (0, 0.09, 0, 0,0,0)
x = cost.keys()
y = cost.values()
total= sum(y)
plt.pie(y,explode=explode,labels=x,autopct=lambda(p): '{:.0f}'.format(p * total/100), shadow=True)
plt.show()
```
