```python
import pickle
from collections import defaultdict
from io import BytesIO
from zipfile import ZipFile
from urllib2 import urlopen
import re
import pandas as pd 
url = urlopen('http://www2.compute.dtu.dk/~faan/data/AFINN.zip')
zipfile = ZipFile(BytesIO(url.read()))
afinn_file = zipfile.open('AFINN/AFINN-111.txt')
```

```python
afinn = dict()
for line in afinn_file:
    parts = line.strip().split()
    if len(parts) == 2:
        afinn[parts[0].decode("utf-8")] = int(parts[1])
```

```python
def afinn_sentiment(terms, afinn):
    total = 0.
    for t in terms:
        if t in afinn:
            #print('\t%s=%d' % (t, afinn[t]))
            total += afinn[t]
    return total
```

```python
def tokenize(text):
    return re.sub('\W+', ' ', text.lower()).split()
```

```python
def manual_labelling(each_dict,excel,team):
    tweet=tokenize(each_dict['tweet'])
    senti=afinn_sentiment(tweet,afinn) 
    if senti <0:
        senti=-1
    else:
        senti=1
    excel=excel.append({'TWEET':each_dict['tweet'].encode("utf-8"),'SENTIMENT':senti, 'TEAM':team},ignore_index=True)
    return excel
```

```python
a=['#brewers.pkl','#cardinals.pkl','#chicagocubs.pkl','#dodgers.pkl','#nationals.pkl','#whitesox.pkl']
excel= pd.DataFrame( columns=['TWEET','SENTIMENT','TEAM'])
for team in a:
    with open(team, "rb") as input_file:
        e = pickle.load(input_file) 
        for each_dict in e:
            excel=manual_labelling(each_dict,excel,team)
```

```python
excel.to_csv('manual_lable',sep='\t',header =True)
```
